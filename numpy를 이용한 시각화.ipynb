## x=0부터5까지 단위1

### y1=x+1

### y2=x+2

### y3=x+3

### y4=x+4

```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r--',marker='.')
plt.plot(x,y2,'b-',marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1','y2'])

plt.subplot(1,2,2)
plt.plot(x,y3,'g',marker='^')
plt.plot(x,y4,'y',marker='*')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y3','y4'])
```
