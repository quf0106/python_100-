```python
s=time.time()
from random import randint
array=[]

for _ in range(1, 10000):
    array.append(randint(1,100))

e=time.time()
print('수행시간 : ',e-s)
```

```python
array
```

```python
s=time.time()
for i in range(len(array)):
    for j in range(len(array)):
        if array[i]<array[j]:
            array[i], array[j] = array[j], array[i]

array
e=time.time()
print('수행시간 : ',e-s)
```
