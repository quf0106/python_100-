```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
text = open(r'nero.txt',encoding='utf-8').read()
text
```

```python
img = r'cat.jfif'
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc = WordCloud(
font_path = 'gulim',
background_color = 'white',
mask = lp_mask,
max_font_size = 50,
contour_width = 5,
contour_color = 'green').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file('1.png')
```
