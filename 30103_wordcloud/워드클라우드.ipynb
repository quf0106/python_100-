```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib 
from matplotlib import rc
rc('font',family='gulim')
```

### 텍스트 가져오기

```python
text = open(r'thelittleprince.txt').read()
text
```

### 이미지 가져오기

```python
img = r'little.png'
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc = WordCloud(
background_color = 'skyBlue',
mask = lp_mask,
max_font_size = 50,
contour_width = 5,
contour_color = 'red').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file('1.png')
```
